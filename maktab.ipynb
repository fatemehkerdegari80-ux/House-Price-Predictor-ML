```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
from sklearn import preprocessing
%matplotlib inline
```

```python
df = pd.read_csv("housePrice.csv")
```

```python
df['Address'].replace('', np.nan, inplace=True)
df.dropna(subset=['Address'], inplace=True)
```

```python
le = preprocessing.LabelEncoder()
le.fit(df["Address"])
address_number=le.transform(df["Address"])
df["address_number"]=address_number
```

```python
df['Area'] = df['Area'].str.replace(',', '')
df['Area'] = df['Area'].astype(float)
df = df[df['Area'] < 2000]
```

```python
df["Parking"]=df["Parking"].replace(True,1)
df["Parking"]=df["Parking"].replace(False,0)
df["Warehouse"]=df["Warehouse"].replace(True,1)
df["Warehouse"]=df["Warehouse"].replace(False,0)
df["Elevator"]=df["Elevator"].replace(True,1)
df["Elevator"]=df["Elevator"].replace(False,0)
```

```python
df["Area"] =df["Area"]/max(df["Area"])
```

```python
msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['Area','Room','Parking','Warehouse','Elevator','address_number']])
y = np.asanyarray(train[['Price']])
regr.fit (x, y)

```

```python

x = np.asanyarray(test[['Area','Room','Parking','Warehouse','Elevator','address_number']])
y = np.asanyarray(test[['Price']])

print('Variance score: %.2f' % regr.score(x, y))
```
